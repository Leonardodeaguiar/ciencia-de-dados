```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from datetime import datetime
from statsmodels.tsa.seasonal import seasonal_decompose
#registro de converters para uso do matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
dados = pd.read_csv("./dados/Airplane_Crashes_and_Fatalities_Since_1908.csv", index_col="Date", parse_dates=True)
dados.resample("D").first()
```

```python
dateparser = lambda x: datetime.strftime(x, "%m-%Y")
```

```python
dados.index
```

```python
dados.drop([ "Time","Location", "Operator", "Flight #", "Type", "Registration", "Summary", "cn/In", "Route"], axis=1, inplace=True)
dados.head()
```

```python
dados.index
```

```python
periodo = dados["Aboard"].loc[ "1995-01-01":"2002-12-31"]
periodo = periodo.resample("M").first()
periodo
```

```python
decomposicao = seasonal_decompose(periodo)
```

```python
plt.plot(decomposicao.trend)
```

```python
plt.plot(decomposicao.resid)
```

```python
plt.plot(decomposicao.seasonal)
```

```python


# Visualização somente da tendência
plt.subplot(4,1,2)
plt.plot(decomposicao.trend, label = 'Tendência')
plt.legend(loc = 'best')

# Visualização somente da sazonalidade
plt.subplot(4,1,3)
plt.plot(decomposicao.seasonal, label = 'Sazonalidade')
plt.legend(loc = 'best')

# Visualização somente do elemento aleatório
plt.subplot(4,1,4)
plt.plot(decomposicao.resid, label = 'Aleatório')
plt.legend(loc = 'best')
plt.tight_layout()
```
